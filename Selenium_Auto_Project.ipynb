```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import logging
```

```python
logging.basicConfig(filename='login_tests.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
```

```python
LOGIN_DATA = [
    {"name": "Gmail", "url": "https://accounts.google.com/signin", "email_field": "identifier", "password_field": "password", "button": "identifierNext", "password_next": "passwordNext"},
    {"name": "Netflix", "url": "https://www.netflix.com/login", "email_field": "id_userLoginId", "password_field": "id_password", "button": "btn-login"},
    {"name": "Hulu", "url": "https://auth.hulu.com/web/login", "email_field": "email_id", "password_field": "password_id", "button": "login-button"},
    {"name": "Spotify", "url": "https://accounts.spotify.com/en/login", "email_field": "login-username", "password_field": "login-password", "button": "login-button"},
    {"name": "Amazon", "url": "https://www.amazon.com/ap/signin", "email_field": "ap_email", "password_field": "ap_password", "button": "signInSubmit"},
    {"name": "YouTube", "url": "https://accounts.google.com/signin/v2/identifier?service=youtube", "email_field": "identifier", "password_field": "password", "button": "identifierNext", "password_next": "passwordNext"},
    {"name": "Instagram", "url": "https://www.instagram.com/accounts/login/", "email_field": "username", "password_field": "password", "button": "loginButton"},
    {"name": "Facebook", "url": "https://www.facebook.com/login/", "email_field": "email", "password_field": "pass", "button": "loginbutton"},
    {"name": "NHL", "url": "https://www.nhl.com/login", "email_field": "username", "password_field": "password", "button": "login-button"},
    {"name": "EliteProspects", "url": "https://www.eliteprospects.com/login", "email_field": "email", "password_field": "password", "button": "login"}
]
```

```python
def test_logins():
    options = Options()
    driver = webdriver.Chrome(options=options)
    results = []
    
    for site in LOGIN_DATA:
        logging.info(f"Testing {site['name']} login...")
        driver.get(site['url'])
        
        try:
            wait = WebDriverWait(driver, 10)
            email_field = wait.until(EC.presence_of_element_located((By.NAME, site['email_field'])))
            email_field.send_keys("dummy@example.com")
            
            # Handle login flow for Gmail and YouTube (email → next → password)
            if "password_next" in site:
                next_button = wait.until(EC.element_to_be_clickable((By.ID, site['button'])))
                next_button.click()
                password_field = wait.until(EC.presence_of_element_located((By.NAME, site['password_field'])))
            else:
                password_field = wait.until(EC.presence_of_element_located((By.NAME, site['password_field'])))
            
            password_field.send_keys("dummyPassword123")
            login_button = wait.until(EC.element_to_be_clickable((By.ID, site['button'])))
            login_button.click()
            results.append((site['name'], "Success: Login button clicked"))
        
        except Exception as e:
            logging.error(f"Failed to test {site['name']}: {str(e)}")
            results.append((site['name'], f"Failed: {str(e)}"))
    
    driver.quit()
    
    logging.info("\nResults:")
    for result in results:
        logging.info(result)
    
    print("\nResults:")
    for result in results:
        print(result)

if __name__ == "__main__":
    test_logins()
```
